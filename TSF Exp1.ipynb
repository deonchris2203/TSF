```python
daily_date = pd.DataFrame(pd.date_range(start=df1['Date'].min(), end=df1['Date'].max()))
daily_date.rename(columns={ daily_date.columns[0]: "Date"}, inplace = True)
daily_date.describe()
```

```python
#ADd missing values
```

```python
df2 = pd.merge(df1,daily_date,on='Date',how='outer')
df2 = df2.sort_values(by=['Date'])
```

```python
df2.describe()
```

```python
df2.head(6)
```

```python
#Linear In==interpolation to fill in null values
print(df2.isnull().sum().sum())
```

```python
df3 = df2.fillna(method='ffill')
print(df3.isnull().sum().sum())
```

```python
df3 = df3.fillna(method='bfill')
print(df3.isnull().sum().sum())
```

```python
df3.head(8)
```

```python
df3.describe()
```

```python
#outlier Detecrtion using Quartile Range
def out_iqr (s, k=1.5, return_thresholds=False):
    #calculate interquartile range
    q25, q75 = np.percentile(s, 25), np.percentile(s, 75)
    iqr = q75 - q25
    #calculate the outlier cutoff
    cut_off = iqr * k
    lower, upper = q25 - cut_off, q75 + cut_off
    if return_thresholds:
        return lower,upper
    else:
        return [True if x < lower or x > upper else False for x in s]
    
    
#For comparision
df4 = df3.drop(columns=['Date'],axis=1)
iqr1 = df4.apply(out_iqr, k=1.5)
```

```python
iqr1.head(10)
```

```python
iqr1.describe()
```

```python
for column in df4:
    df4[column] = np.where(iqr1[column] == True, 'NaN', df4[column])
cols = df4.columns
df4[cols] = df4[cols].apply(pd.to_numeric, errors = 'coerce')
df4.head(10)
```

```python
#use linear interpolation 
df = df4.interpolate(method='linear', axis=0).bfill().ffill()
df3['Date'] = pd.to_datetime(df3['Date'])
df = pd.concat([df3['Date'],df], axis=1)
df.head(10)
```
